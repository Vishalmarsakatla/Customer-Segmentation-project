## Dimensionality Reduction

*Principal Component Analysis (PCA) was used to reduce the dimensionality of the dataset while preserving most of the variance*

**The number of principal components was selected based on cumulative explained variance. PCA components explaining approximately 80% of the variance were retained to balance information preservation and noise reduction**

```python
from  sklearn.decomposition import PCA
pca_full = PCA()
pca_full.fit(scaled_data)

explained_var = pca_full.explained_variance_ratio_
cum_var = np.cumsum(explained_var)

cum_var
```

```python
# Fit PCA on scaled data
pca = PCA()
pca.fit(scaled_data)

# Explained variance
exp_var = pca.explained_variance_ratio_

# Cumulative variance
c_var = np.cumsum(exp_var)

# Create summary table
pca_summary = pd.DataFrame({
    'Principal Component': [f'PC{i+1}' for i in range(len(exp_var))],
    'Explained Variance (%)': exp_var * 100,
   'Cumulative Variance (%)': c_var * 100
})

# Round for readability
pca_summary = pca_summary.round(2)

pca_summary
```

**Based on the explained variance ratio, the first 4 principal components together explain approximately 85% of the total variance in the dataset. This indicates that most of the important information from the original data is retained while significantly reducing dimensionality, making the PCA-transformed data suitable for clustering and visualization.**

```python
from  sklearn.decomposition import PCA
pca = PCA(n_components=4)
df_pca = pca.fit_transform(scaled_data)
pca_df = pd.DataFrame(df_pca, columns=['PC1','PC2','PC3','PC4'])
pca_df.head()
```

## Clustering Techniques

### KMeans Clustering:

```python
from sklearn.cluster import KMeans
```

```python
wcss = []

for i in range(1,13):
    hmodel = KMeans(n_clusters=i)
    hmodel.fit(df_pca)
    wcss.append(hmodel.inertia_)

wcss
```

```python
plt.figure(figsize=(6,3))
plt.plot(range(1,13), wcss,
         marker='H',markerfacecolor='orange',
         linestyle='dotted'
        )

plt.title('Hyperparameter Tuning - Elbow Plot .')
plt.xlabel('No of clusters (k)')
plt.ylabel('WCSS')
plt.show()
```

```python
pca_df
```

```python
final_model = KMeans(n_clusters=4,random_state=42)
final_model.fit_predict(df_pca)
df['Cluster'] =final_model.labels_
```

### DBSCAN = Density-Based Spatial Clustering of Applications with Noise

```python
from sklearn.cluster import DBSCAN
```

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import numpy as np

eps_values = [0.3, 0.5, 0.7, 0.9, 1.1]
min_samples_values = [5, 10, 15, 20]

best_score = -1
best_eps = None
best_min_samples = None

for eps in eps_values:
    for min_s in min_samples_values:

        dbscan = DBSCAN(eps=eps, min_samples=min_s)
        labels = dbscan.fit_predict(scaled_data)

        # remove noise points
        mask = labels != -1

        # need at least 2 clusters
        if len(set(labels[mask])) > 1:
            score = silhouette_score(scaled_data[mask], labels[mask])

            if score > best_score:
                best_score = score
                best_eps = eps
                best_min_samples = min_s

print(f"Optimal DBSCAN parameters: eps={best_eps}, min_samples={best_min_samples}")
print(f"Highest Silhouette Score: {best_score}")
```

```python
optimal_eps = 0.5
optimal_min_samples = 10

dbscan_model = DBSCAN(eps=optimal_eps, min_samples=optimal_min_samples)
dbscan_model.fit(scaled_data)
dbscan_cluster_labels = dbscan_model.labels_

print("DBSCAN clustering applied with optimal parameters.")
print(f"Number of unique clusters found (including noise -1): {len(np.unique(dbscan_cluster_labels))}")
```

```python
df_pca_dbscan = pca_df.copy()
df_pca_dbscan['DBSCAN_Cluster'] = dbscan_cluster_labels

print("df_pca_dbscan created with DBSCAN cluster labels.")
print("First 5 rows of df_pca_dbscan:")
print(df_pca_dbscan.head())
```

```python
plt.figure(figsize=(10, 8))
sns.scatterplot(x='PC1', y='PC2', hue='DBSCAN_Cluster', data=df_pca_dbscan, palette='viridis', s=100, alpha=0.7)
plt.title('DBSCAN Clusters (PCA-reduced Data)')
plt.xlabel('Principal Component 1 (PC1)')
plt.ylabel('Principal Component 2 (PC2)')
plt.legend(title='DBSCAN Cluster')
plt.grid(True)
plt.show()
```

**The DBSCAN visualization shows the presence of noise and irregular cluster shapes; however, the clusters are not clearly separated. This indicates that density-based clustering is less effective for this dataset, and centroid-based methods such as KMeans provide more interpretable segmentation.**

**DBSCAN was evaluated but showed limited performance due to uneven cluster density, making KMeans a more suitable choice for this dataset.**
